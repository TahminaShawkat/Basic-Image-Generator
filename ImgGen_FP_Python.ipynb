```python
#Module
#Libraries
from google.colab.patches import cv2_imshow
from matplotlib import*
import skimage.filters
import cv2
from matplotlib import image              #Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
from matplotlib import pyplot as plt      #pyplot is the submodule,usually importer under the plt alias
import numpy as np 

#Functions
# code for displaying multiple images in one figure 

def flowerList(flowers):
  fig = plt.figure(figsize=(15, 10))                # create figure
  rows = 1                                          # setting values to rows and column variables
  columns = 5
  #rows, colums = 1,5
  # reading images
  fl1 = image.imread('white rose.jpg')
  fl2 = image.imread('1 white-cosmos.jpg')
  fl3 = image.imread('1 white-jasmine.jpg')
  fl4 = image.imread('1 white-tulip.jpg')
  fl5 = image.imread('1 white-orchid.jpg')
  
  fig.add_subplot(rows, columns, 1)                     # Adds a subplot at the 1st position
  plt.imshow(fl1)                                       # showing image
  plt.axis('off')
  plt.title("Rose")
  
  fig.add_subplot(rows, columns, 2)                     # Adds a subplot at the 2nd position
  plt.imshow(fl2)
  plt.axis('off')
  plt.title("Cosmos")
  
  fig.add_subplot(rows, columns, 3)                     # Adds a subplot at the 3rd position
  plt.imshow(fl3)
  plt.axis('off')
  plt.title("Jasmine")
  
  fig.add_subplot(rows, columns, 4)                     # Adds a subplot at the 4th position
  plt.imshow(fl4)
  plt.axis('off')
  plt.title("Tulip")

  fig.add_subplot(rows, columns, 5)                     # Adds a subplot at the 4th position
  plt.imshow(fl5)
  plt.axis('off')
  plt.title("Orchid") 

  plt.show()                                         #plt.show() is showing the pictures after catagory
  plt.waitforbuttonpress(1)

def flower_select(flowers):
  #fig = plt.figure(figsize=(3, 5))                # create figure
  plt.axis('off')
 # reading images

  fl1 = image.imread('white rose.jpg')
  fl2 = image.imread('1 white-cosmos.jpg')
  fl3 = image.imread('1 white-jasmine.jpg')
  fl4 = image.imread('1 white-tulip.jpg')
  fl5 = image.imread('1 white-orchid.jpg')

  if flowers=='Roseüåπ':
    plt.imshow(fl1)
    plt.show()
    return fl1
  elif flowers=='Cosmosüå∫':
    plt.imshow(fl2)
    plt.show()
    return fl2
  elif flowers=='Jasmineüå∏':
    plt.imshow(fl3)
    plt.show()
    return fl3
  elif flowers=='Tulipüå∑':
    plt.imshow(fl4)
    plt.show()
    return fl4
  elif flowers=='Orchidü•Ä':
    plt.imshow(fl5)
    plt.show()
    return fl5
  else:
    print('No flower selected')
  #plt.show()
  #plt.waitforbuttonpress(1)

  #-----------Fruits----------------
def fruitList(fruits) :
  fig = plt.figure(figsize=(15, 10))                # create figure
  rows = 1                                          # setting values to rows and column variables
  columns = 5
# reading images
  fr1 = image.imread('ys banana.jpg')
  fr2 = image.imread('white-orange.jpg')
  fr3 = image.imread('1 white-strawberry.jpg')
  fr4 = image.imread('white apple.jpg')
   
  fig.add_subplot(rows, columns, 1)                    # Adds a subplot at the 1st position
  plt.imshow(fr1)                                    # showing image
  plt.axis('off')
  plt.title("Banana")
    
  fig.add_subplot(rows, columns, 2)                     # Adds a subplot at the 2nd position
  plt.imshow(fr2)
  plt.axis('off')
  plt.title(" Orange")
    
  fig.add_subplot(rows, columns, 3)                     # Adds a subplot at the 3rd position
  plt.imshow(fr3)
  plt.axis('off')
  plt.title("Strawberry")
    
  fig.add_subplot(rows, columns, 4)                     # Adds a subplot at the 4th position
  plt.imshow(fr4)
  plt.axis('off')
  plt.title("Apple")
    
  plt.show()                                            #plt.show() is showing the pictures after catagory
  plt.waitforbuttonpress(1)

def fruit_select(fruits):
  #fig = plt.figure(figsize=(3, 5))                      # create figure
  plt.axis('off')
# reading images
  fr1 = image.imread('ys banana.jpg')
  fr2 = image.imread('white-orange.jpg')
  fr3 = image.imread('1 white-strawberry.jpg')
  fr4 = image.imread('white apple.jpg')
  
  if fruits == 'Bananaüçå':
    plt.imshow(fr1)
    plt.show()
    return fr1
  elif fruits == 'Orangeüçä':
    plt.imshow(fr2)
    plt.show()
    return fr2
  elif fruits == 'Strawberryüçì':
    plt.imshow(fr3)
    plt.show()
    return fr3
  elif fruits == 'Appleüçé':
    plt.imshow(fr4)
    plt.show()
    return fr4
  else:
    print('No fruit selected')
  #plt.show()
  #plt.waitforbuttonpress(1)

#-----------pets----------------
def petList(pets):
  fig = plt.figure(figsize=(15, 10))                # create figure
  rows = 1                                          # setting values to rows and column variables
  columns = 6
  # reading images
  pt1 = image.imread('1 kitty.jfif')
  pt2 = image.imread('2 kitty.webp')
  pt3 = image.imread('3 kitty.jpg')
  pt4 = image.imread('1 puppy.jpg')
  pt5 = image.imread('2 puppy.jpg')
  pt6 = image.imread('3 puppy.jpg')
  
  fig.add_subplot(rows, columns, 1)                    # Adds a subplot at the 1st position
  plt.imshow(pt1)                                    # showing image
  plt.axis('off')
  plt.title('Kitty 1')
  
  fig.add_subplot(rows, columns, 2)                     # Adds a subplot at the 2nd position
  plt.imshow(pt2)
  plt.axis('off')
  plt.title('Kitty 2')
  
  fig.add_subplot(rows, columns, 3)                     # Adds a subplot at the 3rd position
  plt.imshow(pt3)
  plt.axis('off')
  plt.title('Kitty 2')
  
  fig.add_subplot(rows, columns, 4)                     # Adds a subplot at the 4th position
  plt.imshow(pt4)
  plt.axis('off')
  plt.title('Puppy 1')

  fig.add_subplot(rows, columns, 5)                     # Adds a subplot at the 5th position
  plt.imshow(pt5)
  plt.axis('off')
  plt.title('Puppy 2')

  fig.add_subplot(rows, columns, 6)                     # Adds a subplot at the 6th position
  plt.imshow(pt6)
  plt.axis('off')
  plt.title('Puppy 3')
  
  plt.show()                                             #plt.show() is showing the pictures after catagory
  plt.waitforbuttonpress(1)

def pet_select(pets):                                    # this def pet_select(pets): is using to select pets pictures from the album
  #fig = plt.figure(figsize=(3, 5))                      # create figure
  plt.axis('off')
 # reading images
  pt1 = image.imread('1 kitty.jfif')
  pt2 = image.imread('2 kitty.webp')
  pt3 = image.imread('3 kitty.jpg')
  pt4 = image.imread('1 puppy.jpg')
  pt5 = image.imread('2 puppy.jpg')
  pt6 = image.imread('3 puppy.jpg')

  if pets=='Kitty 1':
    plt.imshow(pt1)
    plt.show()
    return pt1
  elif pets == 'Kitty 2':
    plt.imshow(pt2)
    plt.show() 
    return pt2
  elif pets == 'Kitty 3':
    plt.imshow(pt3)
    plt.show()
    return pt3
  elif pets == 'Puppy 1': 
    plt.imshow(pt4)
    plt.show()
    return pt4
  elif pets == 'Puppy 2':
    plt.imshow(pt5) 
    plt.show()
    return pt5
  elif pets == 'Puppy 3':
    plt.imshow(pt6)
    plt.show()
    return pt6
  else:
    print('No pet selected')
  #plt.show()
  #plt.waitforbuttonpress(2)
#-----------colors----------------
def blue(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.0
  img[:,:,0] = img[:,:,0]*0.2
  img[:,:,1] = img[:,:,1]*0.775
  return img

def pink(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.255
  img[:,:,1] = img[:,:,1]*0.105
  img[:,:,2] = img[:,:,2]*0.180
  return img

def orange(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.85
  img[:,:,1] = img[:,:,1]*0.425
  img[:,:,2] = img[:,:,2]*0.150
  return img

def yellow(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.320
  img[:,:,1] = img[:,:,1]*0.650
  img[:,:,2] = img[:,:,2]*0.350
  return img

def red(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.95
  img[:,:,1] = img[:,:,1]*0.1
  img[:,:,2] = img[:,:,2]*0.0
  return img

def purple(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.60
  img[:,:,1] = img[:,:,1]*0
  img[:,:,2] = img[:,:,2]*0
  return img

def green(img1):
  img = img1.copy()
  img[:,:,0] = img[:,:,0]*0.0
  img[:,:,1] = img[:,:,1]*0.900
  img[:,:,2] = img[:,:,2]*0.55
  return img
#-------------EFFECTS----------------
def blur_image(img1):
  img = img1.copy() 
  blur = cv2.blur(img,(15,10))
  imgDisp = np.concatenate((img, blur), axis=1)
  #cv2_imshow(imgDisp)
  cv2.resize(img, (100,100))
  img_45 = cv2.resize(imgDisp, None, fx = 0.45, fy = 0.45)
  cv2_imshow(img_45)
  return img

def pencilSketch_image(img1):
  img = img1.copy()
  s, sketch = cv2.pencilSketch(img, shade_factor = 0.07)
  imgDisp = np.concatenate((img, sketch), axis=1)
  #cv2_imshow(imgDisp)
  cv2.resize(img, (100,100))
  img_45 = cv2.resize(imgDisp, None, fx = 0.45, fy = 0.45)
  cv2_imshow(img_45)
  return img
  
def stylization_image(img1):
  img = img1.copy()
  styled = cv2.stylization(img, sigma_s=100, sigma_r=0.7)
  imgDisp = np.concatenate((img, styled), axis=1)
  #cv2_imshow(imgDisp)
  cv2.resize(img, (100,100))
  img_45 = cv2.resize(imgDisp, None, fx = 0.45, fy = 0.45)                      #img_45 is varialbe where the resized image is saved
  cv2_imshow(img_45)
  return img
```
